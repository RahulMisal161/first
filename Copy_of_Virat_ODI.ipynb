### **5.1 Handling of Inconsitant Data**

```python
df["Runs"] = df["Runs"].str.replace("*", "")
df['Runs']= pd.to_numeric(df['Runs'],errors = 'coerce')
df['Runs']
```

<a name = Section611></a>
### **5.2 Handling of Missing Values**

```python
df['Start Date'] = pd.to_datetime(df['Start Date'], utc=True)
df['Mins']= pd.to_numeric(df['Mins'],errors = 'coerce')
df['Mins'].fillna(df['Mins'].mean(), inplace = True)
df['Mins']
```

```python
df['SR']= pd.to_numeric(df['SR'],errors = 'coerce')
df['SR'].fillna(0, inplace = True)
```
